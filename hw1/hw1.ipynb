```python
from pylab import *
from Bio import SeqIO

#P08100_aaCount.fasta
fileName= input("Please enter the name of the file: ")
windowSize = int(input("Please enter the window: ")) 
file = open(fileName)
for record in SeqIO.parse(file, "fasta"):
 id = record.id
 seq = record.seq
 num_residues = len(seq)
file.close()

hydrophathyScore = { 'A': 1.8,'R':-4.5,'N':-3.5,'D':-3.5,'C': 2.5,
       'Q':-3.5,'E':-3.5,'G':-0.4,'H':-3.2,'I': 4.5,
       'L': 3.8,'K':-3.9,'M': 1.9,'F': 2.8,'P':-1.6,
       'S':-0.8,'T':-0.7,'W':-0.9,'Y':-1.3,'V': 4.2 }

values = []

for residue in seq:
 values.append(hydrophathyScore[residue])

count=0
sumTotal=0
scores=[]
iCounter=0
for i in values:
    iCounter+=1
    if (count == windowSize):
        scores.append(sumTotal/windowSize)
        count=0
        sumTotal=0
    elif (count < windowSize):
        sumTotal= sumTotal + i
        count+=1
        
print(scores)


x_data = range(1, num_residues+1)
plot(x_data, scores, linewidth=1.0)
axis(xmin = 0, xmax = num_residues)
xlabel("Window Size")
ylabel("Average Hydrophobicity")
title("K&D Hydrophobicity for " + id)
show()
```
